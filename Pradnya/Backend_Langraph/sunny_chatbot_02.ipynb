```python
def call_model(state:MessagesState):
    # Messagestate is predefined class-list of messgaes
    messages = state['messages']
    response = llm.invoke(messages)
    return {"messages":[response]}
```

```python
def router_function(state:MessagesState)-> Literal["tools",END]:
    messages = state['messages']
    last_message = messages[-1]
    if last_message.tool_calls:
        return "tools"
    return END
```

```python
# new graph
workflow = StateGraph(MessagesState)
workflow.add_node('agent',call_model)
workflow.add_node("tools",tool_node)

workflow.add_edge(START,'agent')
workflow.add_conditional_edges("agent",router_function,{"tools":"tools",END:END})

app = workflow.compile()
```

5️⃣ SO WHAT IS workflow = mybot() REALLY?
When you write:
workflow = mybot()
Python does this:
Call mybot.__call__()
Build the graph
Compile it
Return a LangGraph app object
Store it in workflow
So now:
workflow
is NOT your chatbot class anymore
It is a LangGraph compiled graph.
That self.app object has methods like:
.invoke()
.stream()

workflow.invoke(...)
means:
“Run the LangGraph workflow with this input.”
“Run the whole chatbot internally and give me only the final answer.”

app.stream(input)
You say:
“Run the chatbot step by step, and after each internal step, tell me what happened.”
So stream() exists to observe the thinking pipeline, not just the final reply.

```python
import sqlite3
conn = sqlite3.connect("demomemory.db", check_same_thread=False)
```

```python
def save_message(conn,role,message):
    conn.execute(
        "INSERT INTO chat_history (role, message) VALUES (?, ?)",
        (role, message)
    )
    conn.commit()
```

```python
from langgraph.checkpoint.sqlite import SqliteSaver

config = {"configurable": {"thread_id": "1"}}

with SqliteSaver.from_conn_string("demomemory.db") as memory:
    app = workflow.compile(checkpointer=memory)

    # runtime loop starts here
    while True:
        user_input = input("User: ")
        if user_input.lower() == "exit":
            
            break

        # save user message
        save_message(conn,"user", user_input)    
        result = app.invoke(
            {"messages": [("user", user_input)]},
            config
        )

        # save bot reply
        save_message(conn,"assistant", result["messages"][-1].content )
        print("Bot:", result["messages"][-1].content)


```
